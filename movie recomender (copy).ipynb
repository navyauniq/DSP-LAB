```python
import ast #AST stands for Abstract Syntax Tree, which is a potent tool of the Python programming language. 
            #It allows us to interact with the Python code itself and can modify it.
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):  #literal_eval function, helps to find the type of value stored in a file
                                         #(some times are known as a node) or strings
        L.append(i['name'])  #getting the value of the name key
    return L 
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['genres'].value_counts()
```

```python
def convert(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ","")) #or else machine will get confused
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
movies['tagline'] = movies['tagline'].apply(lambda x:x.split())
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']+movies['tagline'] 
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew','tagline'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer #sklearn countvectoriz class is used it produces 
                                                            #“spicy sparse matrix” so we converted into numpy array.
                                                            #Convert a collection of text documents to a matrix of token counts.


cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)  
```

```python
recommend('Gandhi')
```

```python
recommend("Pirates of the Caribbean: At World's End")
```

```python
recommend('The Dark Knight Rises')
```
